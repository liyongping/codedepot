```python
%matplotlib inline
import matplotlib.pyplot as plt;
import numpy as np;
import pandas as pd;
import tensorflow as tf;
from sklearn.datasets.samples_generator import make_blobs;
from sklearn.datasets.samples_generator import make_circles;
```

```python
# 人工块数据
centers = [(-2,-2), (-2,1.5), (1.5,-2), (2,1.5)]
data, features = make_blobs(n_samples=100, 
                            centers=centers, 
                            n_features=2, 
                            cluster_std=0.8, 
                            shuffle=False, 
                            random_state=42);
```

```python
plt.scatter(data.T[0], data.T[1], marker='*', c= 'k');
plt.scatter(np.asarray(centers).T[0], 
            np.asarray(centers).T[1], 
            marker='o', 
            c= 'r');
```

```python
data, features = make_circles(n_samples=400, 
                              shuffle=True, 
                              noise=0.1,
                              random_state=None,
                              factor=0.3);

plt.scatter(data.T[0], data.T[1], marker='*', c= 'k');
```
